```python
import math
import numpy as np
```

```python

def calculateOriginalDM(h, N, tm):
    top = h*h*N
    bottom = 4*math.pi*tm
    return math.sqrt(top/bottom )
```

```python
calculateOriginalDM(0.5, 40000, 50)
```

```python
def calculateOriginalHeuristics(h, N, r):
    top = h*h*N
    bottom = 4*math.pi*r*r
    return top/bottom
def piecewiseFunc_3d(h, N, tm, r, ds):
    if(r < ds):
        return 0
    elif (r < calculateOriginalDM(h, N, tm)):
        return tm
    else:
        return calculateOriginalHeuristics(h, N, r)
```

```python
piecewiseFunc_3d(0.5, 40000, 50, 2, 1)
```

```python
def calculateDM(h, N, tm):
    top = h*N
    bottom = 2*math.pi*tm
    return top/bottom
def calculate2dHeuristics(h, N, r):
    top = h*N
    bottom = 2*math.pi*r
    return top/bottom
def piecewiseFunc_2d(h, N, tm, r, ds):
    if(r < ds):
        return 0
    elif (r < calculateDM(h, N, tm)):
        return tm
    else:
        return calculate2dHeuristics(h, N, r)
```

```python
x_list = np.linspace(1,20,400)
```

```python

h = 0.5
N = 1080
tm = 50
ds = 2

print("Limit of second branch " + str(calculateDM(h, N, tm)) )
```

```python
import plotly
import plotly.graph_objs as go
import plotly.io as pio

h = 0.5
N = 1080
tm = 17
ds = 2.5

print("Limit of second branch " + str(calculateDM(h, N, tm)) )

y_list = [piecewiseFunc_2d(h, N, tm, r, ds) for r in x_list]

trace1 = go.Scatter(
    x = x_list,
    y = y_list,
    mode = 'lines',
    name = 'Estimated Measurements'
)
trace2 = go.Scatter(
    x = x_list,
    y = [tm]*400,
    line = dict(color='darkblue', width=4, dash='dot'),
    name = 'Measurement Threshold'
)

trace3 = go.Scatter(
    x = [ds]*400,
    y = range(tm+2),
    name = 'Safety Threshold',
    line = dict(color='limegreen', width=4, dash='dot'))

layout = go.Layout(
    title='Heuristic for surface frontiers',
    titlefont=dict(
        family='Courier New, monospace',
        size=24,
    ),
    xaxis=dict(
        title='Distance from Cell (meters)',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        ),
        dtick=2
    ),
    yaxis=dict(
        title='Estimate Number of Measurements',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        ),
        dtick=2
    )
)


fig=dict(data=[trace1, trace2, trace3], layout=layout)


plotly.offline.plot(fig, filename='./occupied_vs_time.html', image='png')


plotly.offline.iplot(fig)
pio.write_image(fig, '/media/mfaria/Ganesha/20171219_backup_home_catec/Margarida/20190100_exploration/heuristic_2d.pdf', scale=2)
```
